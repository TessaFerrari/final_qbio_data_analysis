```python
import cptac

cptac.download(dataset="Luad")
lung = cptac.Luad()
```

```python
lung.list_data()
```

```python
protein_data = lung.get_proteomics()
protein_data
```

```python
import numpy as np
import pandas as pd
```

```python
rna_data = lung.get_transcriptomics()
clinical_data = lung.get_clinical()
```

```python
print(protein_data.shape)
print(rna_data.shape)
print(clinical_data.shape)
```

```python
name_intersects = [
    np.intersect1d(protein_data.index , rna_data.index) , # 0. fill in intersecting for protein/rna here
    np.intersect1d(protein_data.index , clinical_data.index) , # 1. fill in intersecting for protein/clinical here
    np.intersect1d(clinical_data.index , rna_data.index) , # 2. fill in the intersecting for rna/clinical here
]
for i, intersects in enumerate(name_intersects):
    print(f"The length of {i} is {len(intersects)}")  # fill in here
```

```python
clinical_data
```

```python
sum(pd.isna(clinical_data["Ethnicity"]))
clinical_data["Ethnicity"] = np.where(pd.isna(clinical_data["Ethnicity"]), "unknown", clinical_data["Ethnicity"])
```

```python
np.unique(clinical_data["Ethnicity"])
```

```python
white_mask = (clinical_data["Ethnicity"]=="white") | (clinical_data["Ethnicity"]=="caucasian") | (clinical_data["Ethnicity"]=="european")
black_mask = (clinical_data["Ethnicity"]=="black")
asian_mask = (clinical_data["Ethnicity"]=="asian") | (clinical_data["Ethnicity"]=="han")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
rna_data.columns
```

```python
#protein_data.columns

prot_data_cols= []

for x in protein_data.columns:
    prot_data_cols.append(x[0])
    
prot_data_cols
```

```python
shared_rna_prot = np.intersect1d(rna_data.columns, prot_data_cols)
shared_rna_prot.shape
```

```python
rna_shared_w = rna_data.loc[white_mask,shared_rna_prot]
prot_shared_w = protein_data.loc[white_mask,shared_rna_prot]

rna_shared_b = rna_data.loc[black_mask,shared_rna_prot]
prot_shared_b = protein_data.loc[black_mask,shared_rna_prot]

rna_shared_a = rna_data.loc[asian_mask,shared_rna_prot]
prot_shared_a = protein_data.loc[asian_mask,shared_rna_prot]
```

```python
gene_names = ["TP53", "TTN", "MUC16", "RYR2", "LRP1B", "ZFHX4", "USH2A", "KRAS", "XIRP2"] #genes based on oncoplot
```

```python
# Heatmap for WHITE patients
# NOTE: CSMD3 not included b/c data is not available
corr_df = pd.DataFrame(np.ndarray(shape=(9, 9), dtype=np.float16),
                      index = gene_names,
                      columns = gene_names)

# 2. fill in the data frame!
for g1 in gene_names:
    for g2 in gene_names:
        # calculate the correlations between protein and RNA
        corr, pval = stats.spearmanr(rna_shared_w[g1], prot_shared_w[g2], nan_policy="omit")
        # then, use .loc[] to store the correlation in corr_df
        corr_df.loc[g1, g2] = corr

# 3. create the heat map
sns.heatmap(
    corr_df,
    cmap='mako'
)

plt.show()
```

```python
# Heatmap for BLACK patients
# NOTE: CSMD3 not included b/c data is not available
corr_df = pd.DataFrame(np.ndarray(shape=(9, 9), dtype=np.float16),
                      index = gene_names,
                      columns = gene_names)

# 2. fill in the data frame!
for g1 in gene_names:
    for g2 in gene_names:
        # calculate the correlations between protein and RNA
        corr, pval = stats.spearmanr(rna_shared_b[g1], prot_shared_b[g2], nan_policy="omit")
        # then, use .loc[] to store the correlation in corr_df
        corr_df.loc[g1, g2] = corr

# 3. create the heat map
sns.heatmap(
    corr_df,
    cmap='mako'
)

plt.show()
```
